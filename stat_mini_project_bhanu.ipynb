```python
# 2. A study was done to measure the blood pressure of 60 year old women with glaucoma.
# A random sample of 200 60-year old women with glaucoma was chosen. 
# The mean of the systolic blood pressure in the sample was 140 mm Hg and the standard deviation was 25 mm Hg.
```

```python
n=200
xbar=140
s=25
```

```python
# a)Calculate the estimated standard error of the sample mean? What does the standard error indicate?
```

```python
std_error=s/np.sqrt(n)
std_error                      #std_error indicates that sample mean is 1.77(approx) std deviation away from population mean 
```

```python
# b)Estimate a 95% confidence interval for the true mean blood pressure for all 60-year old women with glaucoma.
```

```python
t_alpha=st.t.isf(0.05/2,n-1)
t_alpha
```

```python
ul_interval=xbar+t_alpha*std_error
ll_interval=xbar-t_alpha*std_error
conf_range=ul_interval-ll_interval
print('range',conf_range,'ul',ul_interval,'ll',ll_interval)
```

```python
# c)Assume that instead of 200, a random sample of only 100 60-year old women with glaucoma was chosen. 
#The sample mean and standard deviation estimates are the same as those in the original study. 
# What is the estimated 95% confidence interval for the true mean blood pressure?
```

```python
n=100
t_alpha=st.t.isf(0.05/2,n-1)
t_alpha
ul_int=xbar+t_alpha*(s/np.sqrt(100))
ll_int=xbar-t_alpha*(s/np.sqrt(100))
conf_range1=ul_int-ll_int
conf_range1                 
print('range',conf_range1,'ul',ul_int,'ll',ll_int)    #it increases when sample size reduces
```

```python
# d) Which of the two above intervals are wider?
```

```python
Second one is wider than the first one as it has less samples
```

```python
#e)Explain in non-technical terms why the estimated standard error of a sample mean tends 
#to decrease with an increase in sample size.
```

```python
std error is deviation of avg sample mean from population mean so as sample size increases  the avg sample mean tends to come
closer to the population mean as no of observation from same sample increases ,so std error decreases
also by formula,std error=std.error/np.sqrt(number of sample values),so as values increases std error decreases
```
